# Model Comparison & Result Discussion

## 1. Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. Enter Evaluation Results

```python
results_dict = {
    "Model": [
        "Linear Regression",
        "SVR",
        "Random Forest",
        "MLP"
    ],
    "RMSE": [
        0.29554600725662683,
        0.11651982662704574,
        0.13948192837705822,
        0.610541104637887
    ],
    "MAE": [
        0.2828176641131316,
        0.09090475941777075,
        0.1145667417448252,
        0.5635074728926631
    ],
    "R2": [
        -9.81304912972446,
        -0.6807288028688261,
        -1.4084291177540167,
        -45.14533459979063
    ]
}

df_results = pd.DataFrame(results_dict)
df_results
```

## 3. Compute Summary Statistics

```python
stats = df_results.set_index("Model").agg(["mean", "std"])
stats
```

## 4. Save Comparison Table

```python
# stats.to_csv("../../results/tables/model_statistics.csv")
stats.to_csv("model_statistics.csv")
```

## 5. Bar Plot – RMSE (Mean ± Std)

```python
plt.figure(figsize=(7, 5))

plt.bar(
    df_results["Model"],
    df_results["RMSE"],
    yerr=df_results["RMSE"].std(),
    capsize=5
)

plt.ylabel("RMSE")
plt.title("Model Comparison: RMSE")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

## 6. Bar Plot – MAE (Mean ± Std)

```python
plt.figure(figsize=(7, 5))

plt.bar(
    df_results["Model"],
    df_results["MAE"],
    yerr=df_results["MAE"].std(),
    capsize=5
)

plt.ylabel("MAE")
plt.title("Model Comparison: MAE")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

## 7. Bar Plot – R² Score

```python
plt.figure(figsize=(7, 5))

plt.bar(
    df_results["Model"],
    df_results["R2"],
    yerr=df_results["R2"].std(),
    capsize=5
)

plt.axhline(0, linestyle="--")
plt.ylabel("R² Score")
plt.title("Model Comparison: R²")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```
